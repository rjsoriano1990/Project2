```python
# importing depencies
import os 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sqlalchemy import create_engine
from config import post_pass
```

```python
#establishing the path to dataset

csv_file = os.path.join("Resources/COVID-19_Vaccinations_in_the_United_States_Jurisdiction.csv")

US_Cov19_df = pd.read_csv(csv_file,low_memory=False)

US_Cov19_df
```

```python
US_Cov19_df.columns.tolist()
```

```python
# Selecting columns of interets

New_US_Cov19_df = US_Cov19_df[['Date',
 'Location',
 'Distributed',
 'Distributed_Janssen',
 'Distributed_Moderna',
 'Distributed_Pfizer','Dist_Per_100K','Administered','Administered_Janssen',
 'Administered_Moderna',
 'Administered_Pfizer','Recip_Administered',
 'Administered_Dose1_Recip',
 'Administered_Dose1_Pop_Pct']].copy()

New_US_Cov19_df
```

```python
# converting the date column format to_date()

New_US_Cov19_df["Date"] = pd.to_datetime(New_US_Cov19_df["Date"])

New_US_Cov19_df
```

```python
# Selecting covid-19 data for the 50 states in the Us

filter_New_US_Cov19_df = New_US_Cov19_df.loc[New_US_Cov19_df ['Location'].isin(["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA", "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"])]

filter_New_US_Cov19_df
```

```python
#export test csv
#final_df.to_csv('test_p.csv', index=False)
```

```python
#rename columns
renamed_US_Cov19_df = filter_New_US_Cov19_df.rename(columns={'Date':"submission_date",
'Location':"state",
 'Distributed':"distributed",
 'Distributed_Janssen':"distributed_janssen",
 'Distributed_Moderna':"distributed_moderna",
 'Distributed_Pfizer':"distributed_pfizer",'Dist_Per_100K':"dist_per_100k",'Administered':"administered",'Administered_Janssen':"administered_janssen",
 'Administered_Moderna':"administered_moderna",
 'Administered_Pfizer':"administered_pfizer",'Recip_Administered':"recip_administered",
 'Administered_Dose1_Recip':"administered_dose1_recip",
 'Administered_Dose1_Pop_Pct':"administered_dose1_pop_pct"})

renamed_US_Cov19_df 
```

```python
#sort and create final df
final_df = renamed_US_Cov19_df.sort_values(by=['state', 'submission_date'])
```
